# Parte I – Resolvendo um Problema Linearmente Separável

```python
#"Decompilando" o dataALL.txt
dadosAll = np.fromfile("./data/dataAll.txt", dtype=np.float64)  # ou outro dtype se souber
print(dadosAll.shape)
```

```python
#Shape dos exemplos do dataAll.txt
exemplosAll = dadosAll.reshape(1000,3)
print(exemplosAll.shape)
print(exemplosAll)
```

Os valores de entrada são as duas coordenadas dos pontos


```python
# Definir a seed para ter reprodutibilidade
np.random.seed(9)



# Função do perceptron
def perceptron(exemplosAll):
    # Inicialização da Matriz de pesos
    n = len(exemplosAll[2])  # tamanho do vetor de pesos
    pesos = np.random.uniform(-0.5, 0.5, size=(n, 1))
    erro = 1
    epocas = 0
    ajuste = 0
    while erro == 1:
        erro = 0
        for x, y, tp in exemplosAll:
            pesos, teste = delta(x, y, pesos, tp)
            if teste == 1:
                ajuste =ajuste + 1
                erro = 1
        epocas= epocas + 1
    print("Pesos finais:")
    print(pesos)
    print("Epocas")
    print(epocas)
    print("Ajustes")
    print(ajuste)

# Treinamento com a condição da função de ativação degrau
def delta(x, y, pesos, tp):
    # Taxa de aprendizagem
    taxa_aprend = 0.1

    # Vetor de entrada com bias
    entrada = np.array([-1, x, y]).reshape(1, -1)  # shape (1, 3)

    # Cálculo do produto escalar
    theta = (entrada @ pesos).item()

    # Função degrau
    if theta < 0:
        theta = 0
    else:
        theta = 1

    # Verifica se precisa ajustar os pesos
    if theta == tp:
        return pesos, 0
    else:
        # Atualiza os pesos
        ajuste = taxa_aprend * (tp - theta) * entrada.T  # shape (3,1)
        return pesos + ajuste, 1

# Chamada do treinamento
perceptron(exemplosAll)
```

```python
# Separar por classe
classe_0 = exemplosAll[exemplosAll[:, 2] == 0][:, :2]
classe_1 = exemplosAll[exemplosAll[:, 2] == 1][:, :2]
# Plot
plt.figure(figsize=(8, 8))
plt.scatter(classe_0[:,0], classe_0[:,1], color='blue', label='Classe 0', edgecolors='k')
plt.scatter(classe_1[:,0], classe_1[:,1], color='red', label='Classe 1', edgecolors='k')

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Pontos classificados")
plt.legend()
plt.grid(True)
plt.show()
```
