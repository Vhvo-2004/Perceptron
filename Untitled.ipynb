# Parte I – Resolvendo um Problema Linearmente Separável

```python
#"Decompilando" o dataALL.txt
dadosAll = np.fromfile("./data/dataAll.txt", dtype=np.float64)  # ou outro dtype se souber
print(dadosAll.shape)
```

```python
#Shape dos exemplos do dataAll.txt
exemplosAll = dadosAll.reshape(1000, 3)

# Convertendo para o formato que a classe Perceptron aceita:
x= exemplosAll[:,:2]
y = exemplosAll[:,2]
print(type(x))
```

Os valores de entrada são as duas coordenadas dos pontos


```python
# Definir a seed para ter reprodutibilidade
np.random.seed(9)

class Perceptron:
    """
    Modelo de Perceptron para classificação binária com aprendizado supervisionado.
    """

    def __init__(self, x, y, taxa_aprend=0.1, epoca_limite=100, trust=False):
        """
        Inicializa o Perceptron com os exemplos de treino e configurações.

        Args:
            x (np.ndarray): Matriz de entrada (amostras x features).
            y (np.ndarray): Vetor de saídas (target).
            taxa_aprend (float): Taxa de aprendizado.
            epoca_limite (int): Limite máximo de épocas.
            trust (bool): Permite continuar indefinidamente até convergir.
        """
        self.x = x
        self.y = y
        self.taxa_aprend = taxa_aprend
        self.epoca_limite = epoca_limite
        self.trust = trust
        self.epocas = 0
        self.ajustes = 0
        self.pesos = None
        self._train()

    def _delta(self, entrada, target):
        entrada = np.insert(entrada, 0, -1)  # adiciona o bias
        entrada = entrada.reshape(1, -1)
        theta = (entrada @ self.pesos).item()
        saida = 1 if theta >= 0 else 0

        if saida == target:
            return self.pesos, 0
        else:
            ajuste = self.taxa_aprend * (target - saida) * entrada.T
            return self.pesos + ajuste, 1

    def _train(self):
        n_amostras, n_features = self.x.shape
        self.pesos = np.random.uniform(-0.5, 0.5, size=(n_features + 1, 1))
        erro = 1

        while erro == 1:
            erro = 0
            for i in range(n_amostras):
                entrada = self.x[i]
                target = int(self.y[i])
                self.pesos, houve_ajuste = self._delta(entrada, target)
                if houve_ajuste:
                    self.ajustes += 1
                    erro = 1
            self.epocas += 1
            if self.epocas > self.epoca_limite and not self.trust:
                erro = 0
                print("O modelo nao convergiu")

    def getParametros(self):
        return self.pesos, self.ajustes, self.epocas

    def getParametrosReta(self):
        pesos = self.pesos.flatten()
        w0 = pesos[0]  # bias (com entrada -1)
        coeficientes = pesos[1:]  # w1 até wn
        w_n = coeficientes[-1]

        if w_n == 0:
            raise ValueError("Não é possível isolar a última variável (peso zero).")

        coef_reta = [-w / w_n for w in coeficientes[:-1]]
        bias = w0 / w_n
        return coef_reta, bias
       
            
        
        
# x e y diretamente do reshape:
x = exemplosAll[:, :2]
y = exemplosAll[:, 2]

# Treinamento
p = Perceptron(x, y, trust=True, epoca_limite=1)

# Parâmetros
print(p.getParametros())
print(p.getParametrosReta())
```

```python
# Obter coeficientes da equação isolando x2
coef, bias = p.getParametrosReta()  # coef = [a], bias = b → x2 = a*x1 + b

# Criar valores para x1 e calcular x2
x_vals = np.linspace(min(exemplosAll[:,0])-1, max(exemplosAll[:,0])+1, 100)
y_vals = coef[0] * x_vals + bias  # x2 = a*x1 + b

# Separar por classe
classe_0 = exemplosAll[exemplosAll[:, 2] == 0][:, :2]
classe_1 = exemplosAll[exemplosAll[:, 2] == 1][:, :2]

# Plot
plt.figure(figsize=(25, 25))
plt.scatter(classe_0[:, 0], classe_0[:, 1], color='blue', label='Classe 0', edgecolors='k')
plt.scatter(classe_1[:, 0], classe_1[:, 1], color='red', label='Classe 1', edgecolors='k')

# Plot da reta de decisão
plt.plot(x_vals, y_vals, color='green', label='Fronteira de decisão', linewidth=1)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Pontos classificados com reta de separação")
plt.legend()
plt.grid(True)
plt.show()
```
